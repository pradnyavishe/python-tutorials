**# Strings in Python - Part 1**

```python
# strings - characters in single, double and triple quotes

name = "Madhavi"       # creating a string
print(name)

print(type(name))      # checking data type

print('Hello World')
print("Hello World")
print("It's easy")

print('''"kw- double Quotes"''')

print("\"kw- double Quotes\"")
```

**# Formatted Strings: 1. Old Style Formatting - % Operator**

```python
name = "Madhavi"
age = 16
print("My name is %s and I'm %d" % (name, age))
# %s, %d are placeholders for the strings and integers.
```

**# 2. str.format() method**

```python
name = "Madhavi"
age = 16
print("My name is {} and I'm {}".format(name, age))
```

```python
# you can reference variables by index or keyword
print("My name is {0} and I'm {1}".format(name, age))
print("My name is {1} and I'm {0}".format(name, age))

print("My name is {name} and I'm {age}".format(name="Keshav", age = 26))
```

**# 3. f-strings**

```python
name = "Pradnya"
age = 28
print(f"My name is {name} and I'm {age}")

print(f"My age after 5 years will be {age+5}")
```

**# Escape Characters - backslash with chars**

```python
print('''"kw-double Quotes"''')

print("\"kw-double Quotes\"")     # double quotes using \

print("\'kw-single Quotes\'")     # single quotes using \

print("Hello\nWorld")             # new line

print("Hello\tWorld")             # tab - space          

```

**# String Operators in Pyhon**

```python
a = "Hello"
b = "Python"

print(a+b)        # concatinate
print(a*2)        # multiple copies   
# [] - slice, [:] - range   -- scroll below

if "H" in a:
    print("Yes")
else:
    print("no")

if "h" not in a:
    print("yes")
else:
    print("no")   

print(r"Hello\nWorld")     # raw string - suppress the escape chars

```

**# String Indexing - Part 2**

```python
my_name = "MADHAVI"
# index:   0123456

print(my_name[0])      # first character of string
print(my_name[1])      # second character of string
print(my_name[2])      # third character of string
print(my_name[3])      # fourth character of string
print(my_name[4])      # fifth character of string
print(my_name[5])      # sixth character of string
print(my_name[6])      # seventh character of string
print(my_name[-1])     # last char from string

name2 =           "Hello World"
# index:           012345678910 
# negative index:  1110987654321 -

print(name2[5])       # blank space is also a char
print(name2[-1])      # last char from string
print(name2[-3])      # 3rd last char from string

```

**# String Slicing**

```python
# syntax: string[start : end : step]

my_name = "MADHAVI"
# index:   0123456

print(my_name[0:3])       # default step = 1
print(my_name[0:3:1])   

print(my_name[0:6:1])

print(my_name[0:6:2])     # step = 2

print(my_name[0:6:3])     # step = 3 

print(my_name[0:6:4])     # step = 4

print(my_name[0:6:5])     # step = 5

print(my_name[3:5:1])

print(my_name[4:6:6])

print(my_name[0:2])       # first 2 char

print(my_name[0:3])       # first 3 char

print(my_name[2:5])       # third to fifth char

print(my_name[1:4])       # second to fourth char

print(my_name[-1:])       # last char of string

print(my_name[6:])        # last char of string

print(my_name[-2:])       # last 2 char of string

print(my_name[-3:])       # last 3 char of string

print(my_name[0: :2])     # every second char

print(my_name[:])         # all char

print(my_name[::])        # all char

print(my_name[::-1])      # reverse the string  
```

**# String Methods**

```python
word = "Hello, Madhavi"

# 1. len()
print(len(word))
```

```python
# 2. upper()
print(word.upper())
```

```python
# 3. lower()
print(word.lower())
```

```python
# 4. count()
print(word.count("a"))
print(word.count("l"))
print(word.count("M"))
```

```python
# 5. find()
print(word.find("e"))
print(word.find("o"))
```

```python
# 6. split()
print(word.split(','))
print(word.split(' '))
```

```python
# 7. replace()
print(word.replace("Madhavi", "Pradnya"))
```

```python
# 8.title()
word2 = "Madhavi is great!"
print(word2.title())
```

```python
# 9. strip()
word2 = "  Hello, World!   "
print(len(word2))
print(word2.strip())
```

```python
# 10. join()
zwords = ("Madhavi", "is", "great!")
print(" ".join(zwords))
print("-".join(zwords))
```
