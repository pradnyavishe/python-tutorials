**#Operators**
**# 1. Arithmetic operators**

```python
a = 5
b = 3
print(a+b)    # addition opertaor
print(a-b)    # substraction opertaor
print(a*b)    # multiplication opertaor
print(a/b)    # division opertaor
print(a%b)    # modulus opertaor
print(a//b)   # floor division opertaor
print(a**b)   # exponentiation opertaor
```

**# 2. Comparison operators - output is boolean value**

```python
a = 5
b = 3
print(a>b)      # greater than operator
print(a<b)      # less than operator
print(a==b)     # equal operator
print(a!=b)     # not equal operator
```

**# 3. Assignment operators**

```python
a = 5 # Assignment operator
```

**# 4. Logical operators**



```python
# Rules for 'and' operator :
# 1. True + True = True
# 2. True + False = False
# 3. False + False = False

a = 10
b = 20 
print(a>10 and b<10)     # and operator
print(a==10 and b==20)
print(a==10 or b<10)     # or operator

# Rule for 'or' operator :
# 1. True + False = True

#  'not' operator
print(not(a==10 and b==20))
```

**# 5. Identity operators - is, is not**

```python
x = [1, 2, 3]
y = x
z = [1, 2, 3]
print(x is y)         # is operator
print(x is z)
print(x is not z)     # is not operator
```

**# 6. Membership operators - in, not in**

```python
my_list = ['apple', 'orange', 'watermelon']
print('apple' in my_list)                     # in operator
print('apple2' in my_list)
print('apple2' not in my_list)                # not in operator   
```

**# 7. Bitwize operators - AND &, OR |, XOR ^, NOT ~, etc**

```python
a = 5                                 # 5 in binary - 0101
b = 3                                 # 3 in binary - 0011
print(a & b)                          # 1 in binary - 0001 

# (0: False, 1: True)

# Rules for '&' operator :
# 1. True + True = True
# 2. True + False = False
# 3. False + False = False
```
