**#Casting in Python**

```python
a = 1
print(type(a))

b = "1"
print(type(b))

c = int(b)
print(type(c))

print(a+c)

print(a+int(b))

# all str type can not be casted into int/numerical type 
name = "Madhavi"
newname = int(name)

# # all int/numerical type can be cast into str type
mynum = 28
mynum2 = str(mynum)
print(type(mynum2))

f1 = 22.56
f2 = int(f1)
print(f2)
print(type(f2))

print(type(int(f1)))

in1 = 26
print(type(float(in1)))
```

**#Implicit Type Casting**

```python
var1 = 10               #int type
var2 = 15.5             #float type       
var3 = (var1+var2)
print(var3)
print(type(var3))
```

```python
bool(0)
bool(1)
```

**#Explicit Type Casting**

```python
int_num = 101
str_num = str(int_num)
print(type(str_num))

a0 = bool(0)
print(a0)
print(type(a0))

a1 = bool(1)
print(a1)
print(type(a1))
```
