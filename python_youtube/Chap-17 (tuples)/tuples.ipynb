**# Tuples in Python**

```python
my_tuple = (1, 2, 3)
print(my_tuple)
print(type(my_tuple))
```

**# Create Tuples: 1. using parenthesis**

```python
my_tuple = (1, 2, 3)
print(my_tuple)
print(type(my_tuple))

my_tuple1 = (1, "Madhavi", True, 3.14)
print(my_tuple1)
print(type(my_tuple1)) 
```

**2. without parenthesis**

```python
tuple1 = 1, 2, 3
print(tuple1)
print(type(tuple1))
```

**3. using the tuple constructor**

```python
# syntax: my_tuple = tuple()

tuple2 = ((1, 2, 9))
print(type(tuple2))
print(tuple2)

list1 = [1, 2, 3]
new_tuple = tuple(list1)
print(new_tuple)
print(type(new_tuple))
```

**4. create a single element**

```python
a = ("a",)        # comma add
print(type(a))
```

**# access tuple - indexing**

```python
fruits = ('apple', 'cherry', 'mango')

print(fruits[0])
print(fruits[-1])
```

**#tupple slicing**

```python
new_tuple = (10, 20, 30, 40, 50)

print(new_tuple[0:3:1])

print(new_tuple[0:3:2])

print(new_tuple[0::3])

print(new_tuple[::-1])
```

**# tuple operations: 1. concatenate - join tuples**

```python
tuple1 = (1, 2, 3)
tuple2 = ('a', 'b')

tuple3 = tuple1 + tuple2
print(tuple3)
```

**2. repetitive**

```python
tuple1 = ('a', 'b')
tuple = ('a', 'b') * 3
print(tuple)

tuple1 = (1, 2, 3, 4)
tuple = (1, 2, 3, 4) * 3
print(tuple)
```

**3. checking a element in a tuple**

```python
tuple = (1, 2, 3)

print(1 in tuple)

print(5 in tuple)
```

**# tuple iteration**

```python
# for loop
fruits = ('apple', 'mango', 'cherry')

for i in fruits:
    print(i)

# while loop
i = 0
while i < len(fruits):
    print(fruits[i])
    i += 1
```

**# tuple methods**

```python
colors = ('blue', 'green', 'orange', 'blue')

# count
print(colors.count('blue'))

# index
print(colors.index('green'))
```

**# tuple functions**

```python
numbers = (2, 3, 1, 5, 6)

# len()
print(len(numbers))

# sum()
print(sum(numbers))

# min()
print(min(numbers))

# max()
print(max(numbers))

# sort
print(sorted(numbers))    # tuple is now list

del tuple
a = sorted(numbers)
numbers_sorted = tuple(a)
print(numbers_sorted)
```

**# tuple pack and unpack**

```python
a = "Madhavi"
b = 21
c = "Engineer"

tuple_pack = a, b, c
print(tuple_pack)
```

```python
name, age, profession = tuple_pack
print(name)
print(age)
print(profession)
```

**# tuple modification**

```python
tuple_numbers = (10, 20, 30)

# tuple_numbers[0] = 100      # error

# how to mutate/modify tuple
list_numbers = list(tuple_numbers)
print(list_numbers)

list_numbers[0] = 100
print(list_numbers)

list_numbers.append(50)
print(list_numbers)

list_numbers.remove(20)
print(list_numbers)
```
