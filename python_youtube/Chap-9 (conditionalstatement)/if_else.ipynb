**#Conditional statement in python - 1. if statement**

```python
# if statement works only for true conditions.

a = 26
b = 108
if b > a:
    print("b is greater than a")

age = 26
if age > 19:
    print("You are an adult")

age = int(input("Enter your age: "))
if age > 19:
    print("You are an adult")



print(10>100)
```

**2. if_else statement**

```python
# else handle false condition 

age = int(input("Enter your age"))
if age > 19:
    print("You are an adult")
else:
    print("You are not an adult")

temperature = 30
if temperature < 25:
    print("It is a cool day")
else:
    print("It is a hot day")    
```

**3. if-elif-else statement**

```python
# multiple conditions

marks = int(input("Enter your marks-100: "))

if marks >= 90:
    print("Garde: A+")
elif marks >= 80:
    print("Grade: A")
elif marks >= 70:
    print("Grade: B")
else:
    print("Grade:C")            
```

**4. nested if-else statment**

```python
# if-else inseide if-else statement
# multiple conditions depend on each other


# Q: positive, negative and zero. positive - even/odd
number = int(input("Enter a number: "))

if number > 0:    # checking positive number
    if number % 2 == 0:
        print("This id a even number")
    else:
        print("This is a odd number")
else:
    if number == 0:
        print("This is zero")
    else:
        print("This is a negative number")
```

**5. conditional expressions (ternary operator)**

```python
age = 26
status = "Major" if age >= 18 else "Minor"
print(status)

age = 16
status = "Major" if age >= 18 else "Minor"
print(status)
```
