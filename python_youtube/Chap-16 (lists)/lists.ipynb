**# Access List - Indexing**

```python
list1 = [10, 20, 30, 40, 50]
# index: 0   1   2   3   4
# index:-5  -4  -3  -2  -1

# first element
print(list1[0])

# second element
print(list1[1])

# last element
print(list1[-1])

# second last element
print(list1[-2])
```

**# List Slicing**

```python
list1 = [10, 20, 30, 40 ,50, 60, 100]
# index:   0   1   2   3   4   5   6
# index:  -7  -6  -5  -4  -3  -2  -1

# slicing syntax
# list_name[start : stop : step]

# start is inclusive, default is 0
# stop is exclusive, default is -1 / last index value
# first 3 elements
print(list1[0:3])

# elements from index 1 to 4
print(list1[1:5])   

# last 3 elements
print(list1[-3:])

# alyernative elements
print(list1[0: :2])          # step is 2

# reverse list
print(list1[::-1])           # step is -1
```

**# List Modifying**

```python
my_list = ['apple', 'banana', 'cherry']

print(my_list)

# replace element at index 1
my_list[1] = "blueberry"
print(my_list)

# add element in list
my_list.append("mango")
print(my_list)

# remove element from the list
my_list.remove("cherry")
print(my_list)
```

**# List Methods**

```python
# 1. append
fruits = ["apple", "banana", "orange"]
fruits.append("cherry")
fruits.append(100)
print(fruits)
```

```python
# 2. extend
fruits = ["apple", "orange"]
more_fruits = ["mango", "cherry"]
fruits.extend(more_fruits)
print(fruits)
```

```python
# 3. insert
fruits = ["apple", "orange"]
fruits.insert(1, "blueberry")
print(fruits)
```

```python
# 4. remove
fruits = ["apple", "banana", "orange", "banana"]
fruits.remove("banana")
print(fruits)
```

```python
# 5. clear
fruits = ["orange", "banana"]
fruits.clear()                   # empty list
print(fruits)
```

```python
# 6. finding index
fruits = ["apple", "banana", "cherry", "banana"]
index = fruits.index("banana")
print(index)

# finding index - within a range
fruits = ["apple", "banana", "cherry", "orange", "banana"]
index = fruits.index("banana", 2)
print(index)
```

```python
# 7. count elements
fruits = ["apple", "banana", "cherry", "orange", "banana"]
count = fruits.count("banana")
print(count)
```

```python
# 8. reverse list
fruits = ["apple", "banana", "cherry"]
fruits.reverse()
print(fruits)
```

```python
# 9. sorting list
numbers = [40, 10, 30, 20]
numbers.sort()               # default sort asc order
print(numbers)

# sorting list in descending order
numbers.sort(reverse=True)
print(numbers)

# sorting strings in a list
fruits = ["cherry", "apple", "banana", "mango", "jackfruit"]
fruits.sort()                                   # default by chars asc order
print(fruits)

# sorting with a key
fruits.sort(key=len)                            # sort based on len default asc order
print(fruits)

fruits.sort(key=len, reverse=True)                            # sort based on len desc order
print(fruits)
```

```python
# 10. pop with index value
numbers = [10, 20, 30, 40]
popped = numbers.pop(2)
print(popped)
print(numbers)

# pop with default - (remove last value)
popped = numbers.pop()
print(popped)
print(numbers)
```

```python
# 11. copying list
fruits = ["apple", "banana", "cherry"]
copy_fruits = fruits.copy()            # shallow copy
print(copy_fruits)

# copying list - modifying the copy does not affect the original list
copy_fruits.append("mango")
print(copy_fruits)
print(fruits)
```

**# List comprehensions**

```python
# syntax:
# list_name = [expressions for item in iterable if condition]

# 3 main components of list comprehension
# expression, for clause, if condition

# create a list of squares
squares = [x**2 for x in range(1, 6)]
print(squares)

# filter even numbers
even_list = [x for x in range(1, 20) if x%2 == 0]
print(even_list)
```

```python
# apply function to each element of a list
my_list = ["apple", "mango", "cherry"]

# my_name = "madhavi"
# print(my_name)
# print(my_name.upper())

print(my_list)
# print(my_list.upper())     # wrong way

uppercase_list = [list.upper() for list in my_list]
print(uppercase_list)
```

```python
# flatten a nested list using list comprehension
nested_list = [[1, 2], [3, 4], [5, 6]]

result = [item for sublist in nested_list for item in sublist]
print(result)

# first, [1, 2] -> 1, 2
# second, [3, 4] -> 3, 4
# third, [5, 6] -> 5, 6

def flatten_list(list):
    return [item for sublist in list for item in sublist]

final_list = flatten_list(nested_list)
print(final_list)
```

**# List Iterations**

```python
fruits = ['apple', 'cherry', 'mango']

# using for loop
for fruit in fruits:
    print(fruit)
```

```python
# using while loop
fruits = ['apple', 'cherry', 'mango']

print("length of list", len(fruits))

index = 0
while index < len(fruits):
    print(fruits[index])
    index += 1
```
