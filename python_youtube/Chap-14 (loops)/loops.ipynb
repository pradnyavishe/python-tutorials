**# Loops in Python - While and For Loop**

**# While Loop**

```python
# while loop

count = 0

while count < 6:        # condition
    print(count)
    count = count + 1
```

```python
count = 5
while count > 0:
    print(count)
    count -= 1
```

**# For Loop**

```python
language = "Python"     # sequence

for x in language:
    print(x)

```

```python
# range function

# range(stop)
# range(start, stop)
# range(start, stop, step)

for i in range(5):              # stop argument
    print(i)

for i in range(5, 10):          # start, stop argument
    print(i)    

for i in range(1, 10, 2):       # start, stop, step argument
    print(i)
```

```python
for i in range(5):
    print(i)
else:
    print("For loop ended")
```

**# Loop Control Statement**

**# 1. Pass Statement**

```python
# for i in range(5):
#     pass

count = 5
while count > 0:
    if count == 3:
        pass
    else:
        print(count)
    count -= 1
```

```python
count = 5
while count > 0:
    if count == 1:
        pass
    else:
        print(count)
    count -= 1
```

**# 2. Break Statement**

```python
for i in range(7):
    if i == 4:
        break
    print(i)
```

**# 3. Continue Statement**

```python
for i in range(7):
    if i == 4:
        continue
    print(i)
```

**# pass vs continue statment**

```python

# don't try - infinite loop
count = 5
while count > 0:
    # print(count)
    if count == 3:
        continue
    else:
        print(count)
    count -= 1
```

```python
# validate user input

while True:
    user_input = input("Enter 'exit' to STOP: ")
    if user_input == 'exit':
        print("Congrats! You guessed it right!")
        break
    print("Sorry, you entered: ", user_input)
```
