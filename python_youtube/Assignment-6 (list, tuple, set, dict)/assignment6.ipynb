**# Q.1: Find the Intersection (common elements) of two lists.**

```python
list1 = [1, 2, 4, 5]
list2 = [4, 5, 6, 7, 8]

# using for loop
def intersection_loop(list1, list2):
    common_list = []
    for item in list1:
        if item in list2 and item not in common_list:
            common_list.append(item)
    return common_list

# print(intersection_loop(list1, list2))  

# using list comprehension
def inetrsection_comp(list1, list2):
    return[item for item in list1 if item in list2]

print(inetrsection_comp(list1, list2))
```

**# Q.2: Find the most frequent element in a list.**
**numbers = [1, 2, 2, 3, 3, 3, 4]**

```python
numbers = [1, 2, 2, 3, 3, 3, 4, 7, 7, 7, 7]

def most_freq(lst):
    max_count = 0
    most_freq = None
    for item in lst:
        count = lst.count(item)
        if count > max_count:
            max_count = count
            most_freq = item
    return most_freq

print(most_freq(numbers))    
```

**# Q.3: Find cumulative sum of a list. numbers = [1, 2, 3, 4]**

```python
numbers = [1, 2, 3, 4]

def cumulative_sum(lst):
    cum_sum = []
    total = 0
    for num in lst:
        total += num
        cum_sum.append(total)
    return cum_sum

print(cumulative_sum(numbers))   

# Using List Comp:
print([sum(numbers[:i + 1]) for i in range(len(numbers))])
```

**# Q.4: Remove duplicates from the list.**
**fruits = ["apple", "banana", "cherry", "apple", "banana"]**

```python
fruits = ["apple", "banana", "cherry", "apple", "banana"]
# using loop
def remove_duplicates(lst):
    unique = []
    seen = set()
    for item in lst:
        if item not in seen:
            unique.append(item)
            seen.add(item)
    return unique

print(remove_duplicates(fruits))

# without seen, but not good for large dataset - list
def remove_duplicates(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique

print(remove_duplicates(fruits))

# using set constructor
print(list(set(fruits)))
```

**# Q.6: Find the most frequent value in a dictionary. data = {'a':1, 'b':2, 'c':1, 'd':3, 'e':1}**

```python
data = {'a':1, 'b':2, 'c':1, 'd':3, 'e':1}

def most_freq(dct):
    frequency = {}
    for value in dct.values():
        if value not in frequency:
            frequency[value] = 0
        frequency[value] += 1      # 1:1, 2:1, 1:2, 3:1, 1:3
    max_value = max(frequency, key=frequency.get)
    return max_value

print(most_freq(data))  

# print(max([1, 2, 6, 8]))
```

**# Q.7: Merge dictionaries with summation.**
**dict1 = {'a':10, 'b':20, 'c'=30}**
**dict2 = {'b':15, 'c':35, 'd':25}**

```python
dict1 = {'a':10, 'b':20, 'c':30}
dict2 = {'b':15, 'c':35, 'd':25}

def merge_dict(dict1, dict2):
    result = dict1.copy()
    for key, value in dict2.items():
        if key in result:
            result[key] += value
        else:
            result[key] = value
    return result

print(merge_dict(dict1, dict2))
```

**# Q.8: Flatten a nested dictionary**
**data = {'a':{'b':{'c':42}, 'd':7}, 'e':10}**

```python
data = {'a':{'b':{'c':42}, 'd':7}, 'e':10}
# o/p {a.b.c:42, a.d:7, e:10}

def flatten_dict(data, parent_key='', sep='.'):
    items = {} #initialize empty dict to store flattened items
    for key, value in data.items():
        # combine current key with parent key
        new_key = f"{parent_key}{sep}{key}" if parent_key else key
        if isinstance(value, dict): # check if dict or not
            # recursive flatten the nested dict
            items.update(flatten_dict(value, new_key, sep))
        else:
            # adding key-value to flatten dict 
            items[new_key] = value
    return items

print(flatten_dict(data))    

print(isinstance(data, dict))
print(isinstance(data, list))
```

**# Q.9: Sort a dictionary by values. data = {'a':5, 'b':9, 'c':2, 'd':7}**

```python
data = {'a':5, 'b':9, 'c':2, 'd':7}

def sort_by_values(data):
    sorted_items = sorted(data.items(),
                          key = lambda item: item[1],
                          reverse=True)
    return {key: value for key, value in sorted_items}

print(sort_by_values(data))

# print(sorted([1, 2, 0, 2, 8], reverse=True))

# print(data.items())
```
