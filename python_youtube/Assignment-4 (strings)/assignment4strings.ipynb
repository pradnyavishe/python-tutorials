**# Q.1: Limit the decimal places to 2 digits using .format mehtod and print result, for the variable pi =     3.14159265359**

```python
pi = 3.14159265359

print(round(pi, 2))

print("Value of pi is {}".format(pi))

# using f-function formating float numbers
print("Value of pi is {:.2f}".format(pi))

print("{:.2f}".format(pi))

# f-strings
print(f"{pi:.2f} using f-string")
```

**# Q.3: Slice to get only the middle character(s): For my_string = "Madhavi", use slicing to extract the middle character(s).**

```python
my_string = "Madhav"                           # 6 chars - even
# index:     012345
my_string2 = "Madhava"                         # 7 chars - odd
# index:      0123456
def mid_str(word):
    middle = int(len(word) / 2)                # 3
    if len(word) % 2 == 0:                     # even char length
        return word[middle-1 : middle+1]       # 2:4    
    else:                                      # odd char length 
        return word[middle]
print(mid_str(my_string))    
print(mid_str(my_string2)) 
```

**# Q.7: Write a Python function to check if a string is a palindrome using string methods.**

```python
# madam = madam    #palindrome

word = "madam"
word2 = "madan"

def is_palindrome(s):
    if s == s[::-1]:
        print(f"{s} is a palindrome")
    else:
        print(f"{s} is not a palindrome")

is_palindrome(word)     
is_palindrome(word2)        
```
